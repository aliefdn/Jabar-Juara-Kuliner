### Data Manipulation (Jumlah & Pendapatan)

```python
# Choosing only sektor wisata column related to culinary and tahun column under or equal 2019
df2_pendapatan = df_pendapatan[(df_pendapatan['sektor_wisata'] == 'RESTORAN/RUMAH MAKAN') &
                               (df_pendapatan['tahun'] <= 2019)]

df2_pendapatan.reset_index(drop=True, inplace=True)                    
```

```python
# Merging between df2_pendapatan and df_usaha_makan. Don't forget to drop the duplicate column.
cols_to_use = df_usaha_makan.columns.difference(df2_pendapatan.columns)

dfNew = df2_pendapatan.merge(df_usaha_makan[cols_to_use],
                             left_index=True,
                             right_index=True,
                             how='inner')

dfNew
```

```python
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 2.5})

fig, ax = plt.subplots(figsize = (17,10))

g = sns.scatterplot(x="jumlah_pendapatan",
                    y="jumlah_resto",
                    data=dfPlot2,
                    hue='nama_kabupaten_kota',
                    legend=True,
                    s=150,
                    palette='Paired')

ax.legend(loc='center left',
          bbox_to_anchor=(1.02, 0.5),
          scatterpoints=1,
          frameon=False,
          labelspacing=1,
          title='Kabupaten/Kota')

xlabels = ['IDR' + '{:,.2f}'.format(x) + 'M' for x in g.get_xticks()/1000000000]
g.set_xticklabels(xlabels)

ax.set_title('Pendapatan dan Jumlah Restoran/Rumah Makan \n Pada Kota/Kabupaten',
             fontdict={'fontsize':24,
                       'fontweight':'bold'})
ax.set_xlabel('Jumlah Pendapatan',
              fontdict={'fontsize':13,
                       'fontweight':'bold'})
ax.set_ylabel('Jumlah Resto',
              fontdict={'fontsize':13,
                       'fontweight':'bold'})

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

sns.despine()

label_point(dfPlot2.jumlah_pendapatan, dfPlot2.jumlah_resto, dfPlot2.nama_kabupaten_kota, plt.gca())
```

### Data Manipulation (Jumlah & Tenaga Kerja)

```python
#Creating groupby and sort the data based on their original kode_kabupaten_kota
df2_usaha_makan = df_usaha_makan.groupby(['nama_kabupaten_kota', 'kode_kabupaten_kota', 'tahun'], as_index=False)['jumlah_resto'].agg('sum')

df2_usaha_makan = df2_usaha_makan.sort_values(['kode_kabupaten_kota', 'tahun'], ascending=True).reset_index(drop=True)

df2_usaha_makan
```

```python
#Creating filter, groupby, and sort value in dataset df_pekerja
df2_pekerja = df_pekerja.groupby(['kode_kabupaten_kota','nama_kabupaten_kota', 'tahun'], as_index=False)['jumlah_tenaga_kerja'].agg('sum')

df2_usaha_makan = df2_usaha_makan.sort_values(['kode_kabupaten_kota', 'tahun'], ascending=True).reset_index(drop=True)

df3_pekerja = df2_pekerja[(df2_pekerja['tahun'] <= 2019)]

df3_pekerja.reset_index(drop=True, inplace=True)

df3_pekerja
```

```python
cols_to_use2 = df3_pekerja.columns.difference(df2_usaha_makan.columns)

df3_pekerja[cols_to_use2]
```

```python
#Merging data between df2_usaha_makan and df3_pekerja
dfNew_set2 = df2_usaha_makan.merge(df3_pekerja[cols_to_use2],
                             left_index=True,
                             right_index=True,
                             how='inner')

dfNew_set2
```

### EDA & Visualization

```python
dfNew_set2.to_csv('data_jumlah_dan_tk_sektor_wisata_kuliner.csv', encoding='utf-8', index=False)
```

```python
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 2.5})

fig, ax = plt.subplots(figsize = (17,10))

ax = sns.lineplot(x='tahun',
                  y='jumlah_tenaga_kerja',
                  data=dfNew_set2,
                  marker='X',)

ax2 = ax.twinx();
sns.lineplot(x='tahun', 
             y='jumlah_resto',
             data=dfNew_set2,
             marker='o',
             color='crimson', lw=3, ax=ax2)


ax.set_title('Jumlah Tenaga Kerja Berdasarkan Restoran/Rumah Makan \n Pada Kota/Kabupaten',
             fontdict={'fontsize':24,
                       'fontweight':'bold'})
ax.set_xlabel('Tahun',
              fontdict={'fontsize':13,
                       'fontweight':'bold'})
ax.set_ylabel('Jumlah Tenaga Kerja',
              fontdict={'fontsize':13,
                       'fontweight':'bold'})
ax2.set_ylabel('Jumlah Resto',
              fontdict={'fontsize':13,
                       'fontweight':'bold'})

ax.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='red')

sns.despine()

plt.show()
```
